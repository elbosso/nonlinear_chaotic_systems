# Exploring the Sprott System M of Differential Equations
## Overview

In this Notebook we explore the Sprott system M of differential equations:

$$
\begin{aligned}
\dot{x} & = -z \\
\dot{y} & = x^2-y \\
\dot{z} & = 1.7+1.7x+y
\end{aligned}
$$

This is one of the many Sprott systems.
